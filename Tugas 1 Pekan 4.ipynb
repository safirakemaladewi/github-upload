Jawab: 
- Machine learning adalah bidang cabang ilmu artificial inteligent (kecerdasan buatan) yang mempelajari tentang bagaimana komputer/machine mampu belajar dari data untuk meningkatkan kecerdasannya, yaitu membuat prediksi terhadap data baru. 
- Data feature merupakan variabel untuk memprediksi suatu kejadian atau target (label), biasanya disebut juga dengan "predictor" -> variabel independent, sedangkan data target adalah variabel dependent atau data yang dipelajari dalam suatu algoritma machine learning.
- Supervised Learning merupakan suatu algoritma machine learning dengan menggunakan data training (data yang sudah berlabel) untuk melakukan pembelajaran. Supervised learning bertujuan agar mesin mampu mengidentifikasi label input yang baru dengan menggunakan fitur yang ada untuk melakukan prediksi maupun klasifikasi. Sedangkan unsupervised learning tidak menggunakan data training dalam melakukan pembelajaran. Sistem mempelajari unlabeled data berdasarkan fitur-fitur dari data tersebut, sehingga tidak memiliki variabel target dan bertujuan mengelompokkan objek yang serupa dalam suatu area tertentu.
- Hyperparameter didefinisikan di awal pembuatan objek model yang bertujuan untuk optimisasi model, sedangkan parameter dihasilkan di akhir, yang didapatkan setelah training data.
- Langkah-langkah dalam mengaplikasikan algoritma apapun dalam machine learning: 1) Pilih model atau algoritma yang akan digunakan. 2) Pilih Hyperparameter ketika membuat objek model. 3) Pisahkan antara data feature dan data target -> sklearn menerima data input bentuk matrik, bukan vektor. 4) Perintahkan model untuk mempelajari data dengan menggunakan method.fit(). 5) Aplikasikan model ke dalam test data dengan menggunakan method predict() untuk supervised learning atau transform() dan predict() untuk unsupervised learning



<h1>Soal 2: Pemahaman Machine Learning</h1>

Pelajarilah secara garis besar suatu model/algoritma machine learning, kemudian aplikasikan untuk membuat model bagi data (variable x, y) di bawah ini. Kemudian buatlah prediksi terhadap data training dan data baru di interval 20-30.

- Plot data asli, data hasil prediksi terhadap data training, data hasil prediksi terhadap data baru
- Tunjukan beberapa parameter yang dimiliki model yang telah kalian buat

```python
import matplotlib.pyplot as plt
import numpy as np

rng = np.random.RandomState(42)
x = 20 * rng.rand(50)
y = x**2 + 2 * x + - 1 + rng.randn(50)

# data yang akan di prediksi
x_new = np.arange(20, 30, 0.5)

fig, ax = plt.subplots(figsize=(12, 6))
ax.scatter(x, y)
ax.set_xlabel('X-Value')
ax.set_ylabel('Y-Value')
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn import tree

model_dtr = DecisionTreeRegressor(criterion='mse', 
                                  splitter='best', 
                                  max_depth=4, 
                                  min_samples_split=2, 
                                  min_samples_leaf=5, 
                                  min_weight_fraction_leaf=0.0,
                                  max_features=None,
                                  random_state=None,  
                                  max_leaf_nodes=7,
                                  min_impurity_decrease=0.0,
                                  min_impurity_split=None,
                                  presort='deprecated',
                                  ccp_alpha=0.0,)

x_train = x.reshape(-1,1)
y_train = y
x_test = x_new.reshape(-1,1)

# perintah model untuk mempelajari data dengan menggunakan method.fit()
model_dtr.fit(x_train, y_train)

y_test = model_dtr.predict(x_test)
y_predict = model_dtr.predict(x_train)

# memvisualisasikan data asli dan prediksi terhadap data baru
fig, ax = plt.subplots(figsize =(12, 8))
plt.scatter(x, y, color='b', label='Data Asli')
plt.plot(x_train, y_predict, color='y', label='Prediksi terhadap Data Training', marker='o', linestyle='None')
plt.plot(x_test, y_test, color='green', label='Prediksi terhadap Data Baru')
plt.legend(bbox_to_anchor=(1.1, 1.2))
plt.show()
tree.plot_tree(model_dtr)
```
